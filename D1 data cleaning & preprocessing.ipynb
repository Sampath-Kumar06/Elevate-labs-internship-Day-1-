**Handling Duplicate data**

```python
data[data.duplicated()]
```

```python
# droping duplicates
data.drop_duplicates(inplace=True)
```

**Handling Null values**

```python
data.isnull().sum()
```

```python
data['Director'].fillna('Not Available', inplace=True)
```

```python
data['Cast'].fillna('Not Available', inplace=True)
```

```python
data['Country'].fillna('Unknown',inplace=True)
```

```python
data['Release_Date'].fillna(data['Release_Date'].mode()[0],inplace=True)
```

```python
data['Rating'].fillna(data['Rating'].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

**Changing data type**

```python
data.head()
```

```python
data.dtypes
```

```python
# changing Release Date data type from object to datatime
data['Release_Date']=pd.to_datetime(data['Release_Date'],format='mixed')
```

```python
# splitting Duration into time and units
data[['duration','unit']]=data['Duration'].str.split(' ',expand=True)
```

```python
data.head()
```

**Standardizing text**

```python
data['Title'] = data['Title'].str.title().str.strip()
data['Category'] = data['Category'].str.lower().str.strip()
```

**Renaming column headers**

```python
columns=[]
for i in data.columns.to_list():
    lower=i.lower()
    columns.append(lower)
```

```python
data.columns=columns
data
```
